### Analysis

- After reviewing the avilable data, Capomulin appears to be an effective treatment regime for SCC tumors. In comparison with other treatment regimes, testing results showed smaller average tumor size, with less variation (standard deviation), between study subjects for those mice treated with Capomulin, with the exception of Ramicane, which had nearly equivalent performance. Of note, Capomulin and Ramicane had greater numbers of study datapoints when compared with the remaining alternate treatement regimes, which may account for the smaller STD DEV and SEM.

- The study also identified that while mouse weight was directly related to tumor size (R^2=0.84), all mice treated with Capomulin showed markedly significant decrease in tumor size over the course of treatement. Specifically, The treatment appears to require some time to take effect, as tumor reduction does not demonstrate reliably until after timpoint= 20 days. After 20 days, the tumor size decreases rapidly.

- Study participants were nearly equally distributed between male and female, however further analysis is required to determine if the efficacy of the treatment is sex-specific. Current data only determines outcomes for pooled study groups.  

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
#study_data_complete = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])
study_data_complete = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Display the data table for preview
study_data_complete.head()
```

```python
# Checking the number of mice.
mouse_count = study_data_complete["Mouse ID"].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_data = study_data_complete.loc[study_data_complete.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()

duplicate_data
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicates = study_data_complete.loc[study_data_complete["Mouse ID"] == "g989"] 
duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_study_data = study_data_complete[study_data_complete["Mouse ID"].isin(duplicate_data)==False]
clean_study_data
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_count = clean_study_data["Mouse ID"].nunique()
clean_mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

# Step 1: Group by Drug Regimen
grouped_data = clean_study_data.groupby("Drug Regimen")

# Step 2: Calculate the mean of the tumor volume for each drug regimen
mean_tumor_volume = grouped_data["Tumor Volume (mm3)"].mean()

# Step 3: Calculate the median of the tumor volume for each drug regimen
median_tumor_volume = grouped_data["Tumor Volume (mm3)"].median()

# Step 4: Calculate the variance of the tumor volume for each drug regimen
variance_tumor_volume = grouped_data["Tumor Volume (mm3)"].var()

# Step 5: Calculate the standard deviation of the tumor volume for each drug regimen
std_dev_tumor_volume = grouped_data["Tumor Volume (mm3)"].std()

# Step 6: Calculate the standard error of the mean (SEM) of the tumor volume for each drug regimen
sem_tumor_volume = grouped_data["Tumor Volume (mm3)"].sem()

# Step 7: Create a DataFrame to display the summary statistics
summary_stats = pd.DataFrame({
    "Mean Tumor Volume": mean_tumor_volume,
    "Median Tumor Volume": median_tumor_volume,
    "Variance Tumor Volume": variance_tumor_volume,
    "Std. Dev. Tumor Volume": std_dev_tumor_volume,
    "SEM Tumor Volume": sem_tumor_volume
})

# Display the summary statistics for each drug regimen
summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

# Group by 'Drug Regimen' and calculate the mean, median, variance, standard deviation, and SEM of the tumor volume
adv_summary_stats = clean_study_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Display the summary statistics for each drug regimen
adv_summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
grouped_data = clean_study_data.groupby("Mouse ID")["Timepoint"].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data = pd.merge(grouped_data, clean_study_data, on=["Mouse ID", "Timepoint"], how="inner")
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
final_tumor_volumes = []

# Locate the rows which contain mice on each drug and get the tumor volumes
for regimen in treatment_regimens:
    tumor_volumes = merged_data.loc[merged_data["Drug Regimen"] == regimen, "Tumor Volume (mm3)"]
    final_tumor_volumes.append(tumor_volumes)

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for i in range(len(treatment_regimens)):
    quartiles = final_tumor_volumes[i].quantile([0.25, 0.5, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q
    lower_bound = lower_q - (1.5 * iqr)
    upper_bound = upper_q + (1.5 * iqr)
    
    # Determine outliers using upper and lower bounds
    outliers = final_tumor_volumes[i].loc[(final_tumor_volumes[i] < lower_bound) | (final_tumor_volumes[i] > upper_bound)]
    
    print(f"Potential outliers for {treatment_regimens[i]}:")
    print(outliers)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

plt.figure(figsize=(10, 6))
box = plt.boxplot(final_tumor_volumes, labels=treatment_regimens, patch_artist=True, 
                  showfliers=True, flierprops=dict(marker='o', markerfacecolor='red', markersize=8))

for b in box["boxes"]:
    b.set(facecolor="none")  

plt.title("Distribution of Tumor Volume by Treatment Group")
plt.xlabel("Treatment Group")
plt.ylabel("Final Tumor Volume (mm3)")
plt.grid(False) 
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

# To match the provided example, use mouse ID = l509
mouse_id = "l509"  

# Filter data for Capomulin regimen
capomulin_data = clean_study_data[clean_study_data["Drug Regimen"] == "Capomulin"] 

# Filter data for the single mouse
single_mouse_data = capomulin_data[capomulin_data["Mouse ID"] == mouse_id]  

# Generate the line plot
plt.figure(figsize=(10, 6))
plt.plot(single_mouse_data["Timepoint"], single_mouse_data["Tumor Volume (mm3)"], marker='o', color='b', label=f"Mouse {mouse_id}")
plt.title(f"Tumor Volume vs. Time Point for Mouse {mouse_id} (Capomulin)")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(False)
plt.show()


```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# Filter data for Capomulin regimen
capomulin_scatter_data = clean_study_data[clean_study_data["Drug Regimen"] == "Capomulin"]  #do I need this again? Or can I reference above?
average_tumor_volume = capomulin_scatter_data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
mouse_weight = capomulin_scatter_data.groupby("Mouse ID")["Weight (g)"].mean()

plt.figure(figsize=(10, 6))
plt.scatter(mouse_weight, average_tumor_volume, color='b', marker='o')
plt.title("Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.grid(False)  
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

from scipy.stats import linregress

# Calculate the correlation coefficient
correlation_coefficient = mouse_weight.corr(average_tumor_volume)

# Fit a linear regression model
slope, intercept, r_value, p_value, std_err = linregress(mouse_weight, average_tumor_volume)

# Print the correlation coefficient
print(f"The correlation between mouse weight and the average tumor volume is: {correlation_coefficient:.2f}")

# Print the linear regression model parameters
#print(f"Linear Regression Model: y = {slope:.2f}x + {intercept:.2f}")

# Generate the scatter plot with the linear regression line
plt.figure(figsize=(10, 6))
plt.scatter(mouse_weight, average_tumor_volume, color='b', marker='o', label="Data Points")
plt.plot(mouse_weight, slope * mouse_weight + intercept, color='r', label="Linear Regression Line")
plt.title("Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.grid(False)
plt.show()
```
